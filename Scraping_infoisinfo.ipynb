```python
## imorting relevant libraries
import requests
from bs4 import BeautifulSoup as bs
```

```python
url = "https://tamil-nadu-state.infoisinfo.co.in/search/construction_company"
request_url = requests.get(url)
request_url
```

```python
## The whole content of the site in an unformatted manner
request_url.content
```

```python
## Using BeautifulSoup to present the html response into a readable format 
soup = bs(request_url.content,'html.parser')
soup
```

```python
soup.find_all('p',class_="address")
##soup.find('p',class_="address").get_text(strip=True)
```

```python
soup.find_all('span',class_="title-com")
## soup.find('span',class_="title-com").get_text(strip=True)
```

```python
page=1
companies = []
while(page != 94):
    basic_url = f"https://tamil-nadu-state.infoisinfo.co.in/search/construction_company/{page}"
    basic_response = requests.get(basic_url)
    basic_soup = bs(basic_response.content,'html.parser')
    for basic in basic_soup.find_all('a',class_="view-company"):
        contractors = {}
        contractors["Name"] = basic.find('span',class_="title-com").get_text(strip=True)
        contractors["Address"] = basic.find('p',class_="address").get_text(strip=True)
        companies.append(contractors)
    page += 1
```

```python
import pandas as pd
import numpy as np
```

```python
len(companies)
```

```python
dataframe = pd.DataFrame(companies)
dataframe
```

```python
filename = "TotalContractor_infoisinfo.xlsx"
excel_file = dataframe.to_excel(filename,encoding="utf-8")
print("success!!!!")
```
